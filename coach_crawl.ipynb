```python
import requests
import bs4
import os
import time
import sys
import time
from urllib.parse import urljoin
from selenium import webdriver
from selenium.webdriver.common.by import By
import chromedriver_autoinstaller


DIV = 'css-1h3fsgr'  # div class name for selection
```

```python
from google.colab import drive
drive.mount('/drive')
```

```python
def download_file(url, dir):
    print(url)
    os.makedirs(dir, exist_ok=True)
    print(f'{dir} was created')
    # setup chrome options
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless') # ensure GUI is off
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')

    # set path to chromedriver as per your configuratlion
    chromedriver_autoinstaller.install()
    driver = webdriver.Chrome(options=chrome_options)  # 把webdriver放在跟py檔同一個料夾, webdriver一定要跟chrome版本一致
    driver.get(url)
    # element = driver.find_element(By.CLASS_NAME, 'variant-image-swatch css-h9wyva')
    elements = driver.find_elements(By.CLASS_NAME, DIV)
    counter = 0
    for element in elements:
        try:
            element.click()
        except:
            print('Download fail: ' + str(url))
            write_file(url)
            continue
        print(element)
        html = driver.page_source
        soup = bs4.BeautifulSoup(html, 'lxml')
        images = soup.find_all('img')
        used_lst = list()  # save all image http routes
        for img_tag in images:
            counter += 1
            http = str(img_tag['src'])
            if not http.startswith('https') or http.startswith('https://bat.bing') or http.startswith(
                    'https://cdn.cookielaw.org/logos/static/powered_by_logo.svg') or http.startswith(
                'https://www.facebook') \
                    or http.startswith('https://cdn.cookielaw.org') or http.endswith('desktopSwatchImage$'):
                continue
            if http.endswith('desktopThumbnail$'):
                http_name = http.split('$')[0]
                http = http_name + '$desktopProduct$'
            if http in used_lst:
                continue
            used_lst.append(http)
            img_url = urljoin(url, http)
            print(img_url)
            img_response = requests.get(img_url)
            file_name = str(counter) + '.jpg'
            img_name = os.path.join(dir, file_name)
            with open(img_name, 'wb') as f:
                f.write(img_response.content)
            print('Download', img_name)


def write_file(url):
    with open('download_fail.txt', 'a') as f:
        f.write(f'{url}\n')
```

```python

input_file = 'coach.txt'
out_dir = '../drive/MyDrive/Download/test'

if os.path.isfile('download_fail.txt'):
  os.remove('download_fail.txt')

if not os.path.isdir('../drive/MyDrive/Download'):
  os.mkdir('../drive/MyDrive/Download')
with open('coach.txt', 'r') as f:
  for i, line in enumerate(f):
    line = line.strip()
    url = line
    dir_ = out_dir+ '_' + str(i)
    download_file(url, dir_)
```
